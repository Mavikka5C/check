```python
import pandas as pd
```

```python
df=pd.read_csv("/content/sales_data_1.csv")
```

```python
df['profit']=(df['Selling price']-df['Buying price'])*df['Quantity sold']
```

```python
df['Firm bought from'].unique()
```

```python
df.groupby('Firm bought from')['profit'].sum()
```

```python
df.groupby('Firm bought from')['profit'].sum().sum() #overall company's profit
```

```python
df.groupby('Firm bought from')['profit'].sum().sort_values(ascending=False)
```

```python
df.groupby('Firm bought from')['profit'].sum().sort_values(ascending=False).tail(1) #least profitable vendor
```

```python
df.groupby('Firm bought from')['profit'].sum().sort_values(ascending=False).head(1) #most profitable vendor
```

```python
df['profit'].sum()*100/(df['Selling price']*df['Quantity sold']).sum()  #gross margin in %
```

```python
df.groupby('Customer')['profit'].sum().sort_values(ascending=False).head(1)  #most profitable customer
```

```python
df.groupby('Date')['profit'].sum()
```

```python
d2=pd.read_csv("/content/sales_data_2.csv")
```

```python
d2
```

```python
d2['profit']=(d2['Selling price']- d2['Buying price'])*d2['Quantity sold']
```

```python
new=pd.DataFrame()
```

```python
new['gm']=d2.groupby('Firm bought from')['profit'].sum()/(d2['Selling price']*d2['Quantity sold']).sum()*100
```

```python
new['qs']=d2.groupby('Firm bought from')['Quantity sold'].sum()/d2.groupby('Firm bought from')['Quantity bought'].sum()*100
```

```python
new
```

```python
new.sort_values(by=['gm','qs'],ascending=False)
```

```python
new.sort_values(by=['gm','qs'],ascending=False).head(1)
```
